```python
import requests
import re
from bs4 import BeautifulSoup
```

```python

url = 'http://www.apple.com'
r = requests.get(url)
r.encoding = 'utf-8'

#Not considering the usage of iframe. 

soup = BeautifulSoup(r.text,'lxml')
raw_links,links = [],[]
for a in soup.find_all('a'):
    link = a['href']
    if link not in raw_links:
        raw_links.append(link)

for link in raw_links:
    print(link)
    if link[0] == '/':
        full_link = url + link
        if full_link != url+'/' and full_link not in links:
            links.append(full_link)
            print(full_link)
    elif link[0] == 'h' and link not in links:
        links.append(link)
        print(link)
print(len(links))
print(len(raw_links))
```

```python
#Not considering the usage of iframe or ajax. 
def find_all_link(url):
    html = requests.get(url)
    html.encoding = 'utf-8'
    soup = BeautifulSoup(html.text,'lxml')
    
    raw_links,links = [],[]
    for a in soup.find_all('a'):
        link = a['href']
        if link not in raw_links and link !='':
            raw_links.append(link)
    for link in raw_links:
        if link[0] == '/':
            full_link = url + link
            if full_link != url+'/' and full_link not in links:
                links.append(full_link)
                print(full_link)
        elif link[0] == 'h' and link not in links:
            links.append(link)
            print(link)
    print(len(links))
```

```python
find_all_link('http://www.apple.com')
```
